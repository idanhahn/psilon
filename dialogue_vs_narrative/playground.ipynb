```python
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
fp = open('../data/Deception Point - Dan Brown.txt', 'r', encoding='utf-8')
data = fp.read()
sentences_raw = tokenizer.tokenize(data)
# print (sentences)
```

```python
sentences = []
x = []
y = []
y_len = []
for idx, sent in enumerate(sentences_raw):
  x.append(idx)
  y_len.append(len(sent))
  if '‘' in sent:
    sentences.append((idx, 1, len(sent), sent))
    y.append(1)
  else:
    sentences.append((idx, 0, len(sent), sent))
    y.append(0)
len(sentences)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 5))
ax = plt.axes()
ax.set_facecolor('red')
plt.bar(x, y, width=0.5, color='blue')
plt.axis('off')
plt.show()
```

```python
gran = len(sentences) / 100
dial = []
narr = []
x = []
for i in range(0, 100):
  x.append(i)
  dial.append(0)
  narr.append(0)
  for sentence in sentences[int(i * gran):int((i + 1) * gran)]:
    if sentence[1] == 1:
      dial[i] += 1
    else:
      narr[i] += 1
print(dial)
print(len(dial))
print(narr)
print(len(narr))

```

```python
sum_dial = sum(dial)
sum_narr = sum(narr)
print(sum_dial)
print(sum_narr)
dial_per = sum_dial / (sum_dial + sum_narr)
print(dial_per)
```

```python
plt.bar(x, narr, width=0.5, color='red')
plt.bar(x, dial, width=0.5,bottom=narr, color='blue')
plt.show()
```
